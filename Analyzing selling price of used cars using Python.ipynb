```python
import pandas as pd
#pandas are powerfull library which are used to data manipulation and data visualization.
import numpy as np
# numpy are used for scientific computing and supports large array or multidimensional array and matrics.
import matplotlib.pyplot as plt
# its a library which is used for data visulaization in python
import seaborn as sns
# seaborn is powerfull data visulaization tool which is  used for high quality graphs 
import scipy as sp
# open source python library which is used for scientific calculation
```

```python
df=pd.read_csv('Book1.csv')
# reading the csv file using the pandas 
df=df.iloc[:,1:]
# iloc is used where first colon represents that all rows are selected,
# refers to selecting all columns starting from the second column (index 1) to the last column.
df.head()
```

```python

df.head()
```

```python
headers=["normalized-losses", "make", 
           "fuel-type", "aspiration","num-of-doors",
           "body-style","drive-wheels", "engine-location",
           "wheel-base","length", "width","height", "curb-weight",
           "engine-type","num-of-cylinders", "engine-size", 
           "fuel-system","bore","stroke", "compression-ratio",
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"]
# headers acontain column names from start all the "column1" etc are named properly
df.columns=headers
# assigning the column to the headers
df.head()
# printing the top 5 rows
```

```python
data=df
data.isna().any()
# to check any NAN values are in the columns
data.isnull().any()
# to check any null values in the column
```

```python
df['city-mpg']=235/df['city-mpg']
# to convert the miles per gallon into kilometer per liter
data.rename(columns = {'city-mpg':'city-L/100km'},inplace=True)
# renaming the column from city_mpg to city-L/100km
print(data.columns)
# to check the change in the column
data.dtypes
# to check the data types
```

```python
data.price.unique()
# to get the unique values in price column
data=data[data.price!='?']

data['price']=data['price'].astype(int)


data.dtypes
```

```python
data['length'].max()
```

```python
data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
data['height'] = data['height']/data['height'].max()
# binning- grouping values
bins = np.linspace(min(data['price']), max(data['price']), 4) 
group_names = ['Low', 'Medium', 'High']
data['price-binned'] = pd.cut(data['price'], bins, 
                              labels = group_names, 
                              include_lowest = True)

print(data['price-binned'])
plt.hist(data['price-binned'])
plt.show()
```

```python
pd.get_dummies(data['fuel-type']).head()
# get dummies is used to get dummy value
data.describe()
```

```python
# examples of box plot
plt.boxplot(data['price'])

# by using seaborn
sns.boxplot(x ='drive-wheels', y ='price', data = data)

# Predicting price based on engine size
# Known on x and predictable on y
plt.scatter(data['engine-size'], data['price'])
plt.title('Scatterplot of Enginesize vs Price')
plt.xlabel('Engine size')
plt.ylabel('Price')
plt.grid()
plt.show()
```

```python
# Grouping Data
test = data[['drive-wheels', 'body-style', 'price']]
data_grp = test.groupby(['drive-wheels', 'body-style'], 
                         as_index = False).mean()

data_grp
```

```python
# pivot method
data_pivot = data_grp.pivot(index = 'drive-wheels',
                            columns = 'body-style')
data_pivot

# heatmap for visualizing data
plt.pcolor(data_pivot, cmap ='RdBu')
plt.colorbar()
plt.show()
```

```python
# Analysis of Variance- ANOVA
# returns f-test and p-value
# f-test = variance between sample group means divided by 
# variation within sample group
# p-value = confidence degree
data_annova = data[['make', 'price']]
grouped_annova = data_annova.groupby(['make'])
annova_results_l = sp.stats.f_oneway(
                             grouped_annova.get_group('honda')['price'],
                             grouped_annova.get_group('subaru')['price']
                                    )
print(annova_results_l)

# strong corealtion between a categorical variable
# if annova test gives large f-test and small p-value

# Correlation- measures dependency, not causation
sns.regplot(x ='engine-size', y ='price', data = data)
plt.ylim(0, )
```
